```python
import pandas as pd

df_banapresso = pd.read_csv('/content/drive/MyDrive/AI활용 소프트웨어 개발/11_데이터분석/data/banapresso_address.csv')
df_banapresso
```

```python
df = df_banapresso.copy()
df
```

```python
df.isna().sum()
```

```python
import folium
```

```python
# 지도 생성
map = folium.Map(location=[df['latitude'].iloc[0], df['longitude'].iloc[0]], zoom_start=12)

# 마커 추가
for _, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['store'],
        tooltip=row['store']
    ).add_to(map)

# HTML로 저장
map.save('banapresso_map.html')
```
